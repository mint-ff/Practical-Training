```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib import rcParams

#  Matplotlib 
rcParams['font.sans-serif'] = ['SimHei']  # 
rcParams['axes.unicode_minus'] = False  # 

# 
G = nx.DiGraph()

# 
# 
G.add_node(r"$x_1$1", pos=(0.5, 6))
G.add_node(r"$x_2$2", pos=(1.5, 6))

#  -  x1 2x2 + 1
G.add_node(r"$x_1^2$", pos=(0.5, 5))
G.add_node(r"$2x_2$2", pos=(1.5, 5))
G.add_node(r"$z_1=x_1^2 + 2x_2 + 1$", pos=(1, 4))

# Sigmoid 
G.add_node(r"$s_1 = \sigma(x_1^2 + 2x_2 + 1)$Sigmoid", pos=(1, 3))

# 
G.add_node(r"$z_2 = 4s_1 + 1$", pos=(1, 2))

# 
G.add_node(r"$y = \sigma(z_2)$", pos=(1, 1))

# 
# 
G.add_edge(r"$x_1$1", r"$x_1^2$")
G.add_edge(r"$x_2$2", r"$2x_2$2")
G.add_edge(r"$x_1^2$", r"$z_1=x_1^2 + 2x_2 + 1$")
G.add_edge(r"$2x_2$2", r"$z_1=x_1^2 + 2x_2 + 1$")

#  Sigmoid 
G.add_edge(r"$z_1=x_1^2 + 2x_2 + 1$", r"$s_1 = \sigma(x_1^2 + 2x_2 + 1)$Sigmoid")

# Sigmoid 
G.add_edge(r"$s_1 = \sigma(x_1^2 + 2x_2 + 1)$Sigmoid", r"$z_2 = 4s_1 + 1$")

# 
G.add_edge(r"$z_2 = 4s_1 + 1$", r"$y = \sigma(z_2)$")

# 
pos = nx.get_node_attributes(G, 'pos')

# 
plt.figure(figsize=(12, 10))
nx.draw(
    G, pos, with_labels=True, node_size=3500, 
    node_color="lightblue", font_size=10, font_weight="bold", arrowsize=20,
    edge_color="black"
)

# 
plt.title("", fontsize=16)
plt.show()
```

```python
import torch

#  x1  x2
x1 = torch.tensor(0.0, requires_grad=True)
x2 = torch.tensor(1.0, requires_grad=True)

# 
z1 = x1**2 + 2*x2 + 1
s1 = torch.sigmoid(z1)
z2 = 3 * s1 + s1 + 1  #  s2 s2 = s1
y = torch.sigmoid(z2)

# 
y.backward(retain_graph=True)  #  x1  x2 

#  z1/x1, s1/x1, z2/x1, y/x1
print("\n x1:")
dz1_dx1 = 2 * x1.item()
print(f"z1/x1 = 2 * x1 = {dz1_dx1}")

ds1_dz1 = s1.item() * (1 - s1.item())
ds1_dx1 = ds1_dz1 * dz1_dx1
print(f"s1/x1 = s1/z1 * z1/x1 = {ds1_dz1} * {dz1_dx1} = {ds1_dx1}")

dz2_ds1 = 3 + 1  # z2  s1 
dz2_dx1 = dz2_ds1 * ds1_dx1
print(f"z2/x1 = z2/s1 * s1/x1 = {dz2_ds1} * {ds1_dx1} = {dz2_dx1}")

dy_dz2 = y.item() * (1 - y.item())
dy_dx1 = dy_dz2 * dz2_dx1
print(f"y/x1 = y/z2 * z2/x1 = {dy_dz2} * {dz2_dx1} = {dy_dx1}")

#  z1/x2, s1/x2, z2/x2, y/x2
print("\n x2:")
dz1_dx2 = 2
print(f"z1/x2 = 2")

ds1_dx2 = ds1_dz1 * dz1_dx2
print(f"s1/x2 = s1/z1 * z1/x2 = {ds1_dz1} * {dz1_dx2} = {ds1_dx2}")

dz2_dx2 = dz2_ds1 * ds1_dx2
print(f"z2/x2 = z2/s1 * s1/x2 = {dz2_ds1} * {ds1_dx2} = {dz2_dx2}")

dy_dx2 = dy_dz2 * dz2_dx2
print(f"y/x2 = y/z2 * z2/x2 = {dy_dz2} * {dz2_dx2} = {dy_dx2}")
```

```python
import torch
import torch.nn.functional as F

#  x1  x2
x1 = torch.tensor(0.0, requires_grad=True)
x2 = torch.tensor(1.0, requires_grad=True)

# 
z1 = x1**2 + 2*x2 + 1
print(f"z1 = {z1.item()} (z1  x1  x2 z1 = x1^2 + 2 * x2 + 1)")

s1 = torch.sigmoid(z1)
s2 = torch.sigmoid(z1)
print(f"s1 = {s1.item()} (s1  z1  Sigmoids1 = sigmoid(z1))")
print(f"s2 = {s2.item()} (s2  z1  Sigmoids2 = sigmoid(z1))")

z2 = 3 * s1 + s2 + 1
print(f"z2 = {z2.item()} (z2 = 3 * s1 + s2 + 1)")

y = torch.sigmoid(z2)
print(f"y = {y.item()} (y  z2  Sigmoidy = sigmoid(z2))")

#  x1  x2 

#  x1 
y.backward(retain_graph=True)  # 

#  x1  x2 
dy_dx1 = x1.grad.item()  #  x1 
dy_dx2 = x2.grad.item()  #  x2 

# 
print(f"\n---------  ---------")
print(f" x1 ")
print(f"1.  z1 = x1^2 + 2 * x2 + 1  x1 ")
print(f"    z1/x1 = 2 * x1 = 2 * {x1.item()} = {2 * x1.item()}")
print(f"2.  s1 = sigmoid(z1)  z1 ")
print(f"    s1/z1 = s1 * (1 - s1) = {s1.item()} * (1 - {s1.item()}) = {s1.item() * (1 - s1.item())}")
print(f"3.  z2 = 3 * s1 + s2 + 1  s1 ")
print(f"    z2/s1 = 3")
print(f"4.  y = sigmoid(z2)  z2 ")
print(f"    y/z2 = y * (1 - y) = {y.item()} * (1 - {y.item()}) = {y.item() * (1 - y.item())}")
print(f"5.  x1 ")
print(f"    y/x1 = y/z2 * z2/s1 * s1/z1 * z1/x1 = {dy_dx1}")
print(f"--------------------------------")

print(f"\n x2 ")
print(f"1.  z1 = x1^2 + 2 * x2 + 1  x2 ")
print(f"    z1/x2 = 2")
print(f"2.  s1 = sigmoid(z1)  z1 ")
print(f"    s1/z1 = s1 * (1 - s1) = {s1.item()} * (1 - {s1.item()}) = {s1.item() * (1 - s1.item())}")
print(f"3.  z2 = 3 * s1 + s2 + 1  s1 ")
print(f"    z2/s1 = 3")
print(f"4.  y = sigmoid(z2)  z2 ")
print(f"    y/z2 = y * (1 - y) = {y.item()} * (1 - {y.item()}) = {y.item() * (1 - y.item())}")
print(f"5.  x2 ")
print(f"    y/x2 = y/z2 * z2/s1 * s1/z1 * z1/x2 = {dy_dx2}")
print(f"--------------------------------")

# 
print(f"\ndy/dx1 = {dy_dx1} ( x1 )")
print(f"dy/dx2 = {dy_dx2} ( x2 )")
```
